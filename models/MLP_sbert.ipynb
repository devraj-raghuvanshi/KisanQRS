```python
def get_f1(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```

```python
def generate_ngrams(sent):
  ngrams_list = []
  for i in range(1,7):
    if i > len(word_tokenize(sent)):
      break
    else:
      i_grams = ngrams(sent.split(), i)
      for grams in i_grams:
        ngrams_list.append(' '.join(grams))
  return ngrams_list
```

```python
def remove_special_char(string):
  special_characters=[',', '?', '-', '.']
  normal_string = "".join(filter(lambda char: char not in special_characters , string))
  return normal_string
```

```python
remove_special_char('varieties of tomato, onions and cucumber')
```

```python
def clean_query(query):
  query = remove_special_char(query)

  new_sent = []
  crop_list = []
  for item in generate_ngrams(query):
    if ps.stem(item) in crops:
      crop_list.append(ps.stem(item))

  crop_list_word_tokens = []
  for crp in crop_list:
    crop_list_word_tokens += word_tokenize(crp)
  for i in range(len(crop_list_word_tokens)):
    crop_list_word_tokens[i] = ps.stem(crop_list_word_tokens[i])

  words_list = word_tokenize(remove_stopwords(query))
  for word in words_list:
    if ps.stem(word) not in crop_list_word_tokens:
      new_sent.append(word)
  return ' '.join(new_sent), crop_list
```

```python
def preprocess_str(str):
  str = str.lower()
  str = str.replace('-', ' ')
  str = remove_special_char(str)

  word_list = []
  for word in word_tokenize(str):
    word_list.append(ps.stem(word))
  return ' '.join(word_list)
```

```python
def cal_str_similarity(s1, s2):

  s1 = preprocess_str(s1)
  s2 = preprocess_str(s2)

  set1 = set(s1.split())
  set2 = set(s2.split())

  set_12_union = set1.union(set2)
  set_12_intersection = set1.intersection(set2)

  if len(set_12_union) == 0:
    return 0
  return len(set_12_intersection)/len(set_12_union)

def cal_char_similarity(s1, s2):
  s1 = s1.lower()
  s2 = s2.lower()
  letters1 = [x for x in s1 if (x not in [' ', ',', '-','.'])]
  letters2 = [x for x in s2 if (x not in [' ', ',', '-','.'])]

  set1 = set(letters1)
  set2 = set(letters2)

  set_12_union = set1.union(set2)
  set_12_intersection = set1.intersection(set2)

  if len(set_12_union) == 0:
    return 0
  
  return len(set_12_intersection)/len(set_12_union)
```

```python
def generate_simil_ans(df, scores, ques_idx, num):
  index_list = [index for index in range(len(scores))]
  sim_ques_idx_list = []
  for tup in sorted(zip(scores[ques_idx], index_list), reverse=True)[1:num+1]:
    idx = tup[1]
    sim_ques_idx_list.append(idx)

  sim_queries = df.loc[sim_ques_idx_list]
  return sim_queries
```

```python
def make_ans_clusters(query):
  df_list = []
  for df in generate_answers(query):
    if df.empty:
      df_list.append(pd.DataFrame())
      continue

    answers = [ans for ans in df.KccAns]
    scores = df.KccAns.apply(lambda x: [(cal_char_similarity(x, answers[idx]) + cal_str_similarity(x, answers[idx]))/2 for idx in range(len(answers))])
    scores = np.array(scores.tolist())

    idx_checklist = []
    clust_list = []

    for i in range(len(df)):
      ques_idx_i = df.iloc[i,:]['index']
      if ques_idx_i not in idx_checklist:
        simil_ques_list = generate_simil_ans(df, scores, i, len(df.KccAns) - 1)['index'].to_list()
        clust = []
        clust.append(ques_idx_i)
        idx_checklist.append(ques_idx_i) 

        count = 0

        for que in simil_ques_list:
          if que not in idx_checklist:
            que_idx = df[df['index'] == que].index[0]
            if scores[i][que_idx] >= 0.4:                                           #threshold
              clust.append(que)
              idx_checklist.append(que)  
        clust_list.append(clust)

    df_reset_idx = df.set_index('index')
    df_reset_idx['AnsCluster'] = np.zeros(len(df_reset_idx['KccAns']))

    clust_num = 1

    df_clust = pd.DataFrame()

    for clu in clust_list:
      if len(clu) >= 1:
        temp1 = df_reset_idx.loc[clu]
        temp1['AnsCluster'] = [clust_num] * len(clu)
        df_clust = pd.concat([df_clust, temp1])
        clust_num += 1
      else:
        temp2 = df_reset_idx.loc[clu]
        temp2['AnsCluster'] = [-1] * len(clu)
        df_clust = pd.concat([df_clust, temp2])

    df_clust.AnsCluster = df_clust.AnsCluster.astype(int)

    df_list.append(df_clust)

  return df_list
```
