```python
def ans_clean(ans):
  ans = str(ans)
  token_list = word_tokenize(ans)
  return ' '.join(token_list)
```

## Removing rows containing
### 1) '-', '--', ':' etc in ans
### 2) 'weather' in query

```python
def remove_special_char(string):
  special_characters=['-',':']
  normal_string = "".join(filter(lambda char: char not in special_characters , string))
  return normal_string


def not_contains_only_special_char(sent):
  new_sent = remove_special_char(sent)
  return bool(new_sent)
```

```python
not_contains_only_special_char(remove_special_char('----'))
```

```python
df_up_150000_rem_spec_char = df_up_150000[df_up_150000['KccAns'].apply(not_contains_only_special_char)]
```

```python
def not_contains_weather(sent):
  sent = sent.lower()
  if 'weather' in word_tokenize(sent):
    return False
  return True
```

```python
df_up_150000_rem_weather = df_up_150000_rem_spec_char[df_up_150000_rem_spec_char['QueryText'].apply(not_contains_weather)]
```

```python
def remove_words(sent):
  sent = sent.lower()
  words = ['recommended for ', 'recommended ']
  for word in words:
    if word in sent:
      return sent.replace(word, '')
  return sent

def remove_words2(sent):
  sent = sent.lower()
  words = ['farmer want to know information about ', 'farmer wants to know information sbout ', 'farmer needs information regarding', 'farmer needs information ', 'farmer need information ','farmer need information regarding', 'farmer want to know information ', 'farmer wants to know information ', 'asked about', 'asking about']
  for word in words:
    if word in sent:
      return sent.replace(word, '')
  return sent

print(remove_words('Recommended for spray profenophose 2 ml Teepol...'))
```

```python
df_up_150000_rem_weather.KccAns = df_up_150000_rem_weather.KccAns.apply(remove_words)
df_up_150000_rem_weather.QueryText = df_up_150000_rem_weather.QueryText.apply(remove_words2)
```
